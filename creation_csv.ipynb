```python
import numpy as np
import pandas as pd
```

```python
df_communes = pd.read_csv("donnees_communes.csv")
df_communes_sorted = df_communes.sort_values(by=['COM'])
df_communes_sorted.reset_index(drop=True, inplace = True)
df_communes_sorted["index"] =  df_communes_sorted.index + 1
df_communes_sorted
```

```python
df_population = df_communes_sorted[["REG","CODREG","COM","PMUN","PCAP","PTOT"]]
df_population
```

```python
df_geographie = pd.read_csv("geographie.csv")
df_geographie_sorted = df_geographie.sort_values(by=['com_nom_maj_court'])
df_geographie_sorted.reset_index(drop=True, inplace = True)
df_geographie_sorted["index"] =  df_geographie_sorted.index + 1
df_geographie_sorted
```

```python
df_geographie = df_geographie[["reg_nom","dep_nom","com_code","com_nom_maj_court"]]
df_region = df_geographie[["reg_nom","dep_nom","com_code","com_nom_maj_court"]]
```

```python
df_valeur_fonciere = pd.read_csv("valeur_fonciere.csv", low_memory=False)
df_valeur_fonciere
df_code_postal = df_valeur_fonciere[["Commune","Code postal"]]
df_code_postal = df_code_postal.drop_duplicates(subset=['Commune'])
df_code_postal = df_code_postal.sort_values(by=['Commune'])
df_code_postal.reset_index(drop=True, inplace = True)
df_code_postal["index"] =  df_code_postal.index + 1
df_code_postal
```

```python
df_region = df_region.drop_duplicates(subset=['reg_nom'])
df_region.reset_index(drop=True, inplace = True)
df_region["id_reg"] = df_region.index + 1
df_region = df_region[["id_reg","reg_nom"]]
df_region.to_csv("region.csv")
```

```python
df_dep = df_geographie
df_dep = df_region.merge(df_dep)
df_dep = df_dep.drop_duplicates(subset=['dep_nom'])
df_dep.reset_index(drop=True, inplace = True)
df_dep["dep_code"] = df_dep.index + 1
df_dep = df_region.merge(df_dep)
df_dep = df_dep[["id_reg","dep_nom","dep_code"]]
df_dep.to_csv("dep.csv")
df_dep
```

```python
df_communes = df_geographie_sorted.merge(df_communes_sorted)
df_communes = df_communes[['reg_nom','dep_nom','com_code','com_nom_maj_court']]
df_communes = df_communes.merge(df_region)
df_communes = df_communes.merge(df_dep)
df_communes= df_communes[["dep_code",'com_code','com_nom_maj_court']]
df_communes =df_communes.rename(columns={"dep_code": "code_departement", "com_code": "code_commune", "com_nom_maj_court" : "Commune"})
df_communes = pd.merge(df_communes, df_code_postal, how="outer", on=["Commune"])
df_communes.drop_duplicates("Commune", inplace = True)
df_communes = df_communes.sort_values(by=['Commune'])
df_communes = df_communes =df_communes.rename(columns={"Commune" : "nom_commune"})

df_communes.drop(columns = "index",inplace = True)
df_communes.reset_index(drop=True, inplace = True)
df_communes["id_codedep_codecommune"] = df_communes.index + 1
df_communes.to_csv("communes.csv")
df_communes
```

```python
df_population = df_communes_sorted
df_population.drop_duplicates("COM",inplace = True)
df_population = df_population[["PMUN","PCAP","PTOT"]]
df_population.reset_index(drop=True, inplace = True)
df_population["id_codedep_codecommune"] = df_population.index + 1
df_population = pd.merge(df_communes, df_population, how="inner", on=["id_codedep_codecommune"])
df_population = df_population[["id_codedep_codecommune","PMUN","PCAP","PTOT"]]
df_population.to_csv("population.csv")
```

```python
df_valeur_fonciere = df_valeur_fonciere[["No voie","Code postal", "Type de voie", "Voie", 
                                        "Nombre pieces principales", "Surface reelle bati","Surface Carrez du 1er lot",
                                       "Type local"]]
df_valeur_fonciere.rename(columns = {"No voie" : "no_voie", "Type de voie" : "type_voie", "Voie" : "voie",
                        "Nombre pieces principales" : "total_piece", "Surface reelle bati": "surface_carrez",
                         "Surface Carrez du 1er lot" : "surface_local", "Type local": "type_local"}, inplace = True)
```

```python
df_valeur_fonciere["index"] = df_valeur_fonciere.index + 1
df_valeur_fonciere
```

```python
df_valeur_fonciere
df_valeur_fonciere = pd.merge(df_valeur_fonciere, df_communes, how="inner", on=["Code postal"])
df_valeur_fonciere
```

```python
df_valeur_fonciere.drop_duplicates("index",inplace = True)
df_valeur_fonciere
```

```python
df_valeur_fonciere = df_valeur_fonciere[["no_voie","type_voie","voie","total_piece","surface_carrez","surface_local","type_local",
                                        "id_codedep_codecommune"]]

df_valeur_fonciere.reset_index(drop=True, inplace = True)
df_valeur_fonciere["id_bien"] = df_valeur_fonciere.index + 1
df_valeur_fonciere.to_csv("valeur_fonciere.csv")
df_valeur_fonciere
```
